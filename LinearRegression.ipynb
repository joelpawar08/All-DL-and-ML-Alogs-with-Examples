```python
# Import necessary libraries
# numpy for numerical operations, matplotlib for plotting, sklearn for machine learning models
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Step 1: Create a self-made dataset
# We'll simulate house prices (y) based on house size in square feet (x)
# Dataset: House sizes (1000 to 3500 sq ft) and prices (in $1000s)
house_sizes = np.array([[1000], [1200], [1500], [1800], [2000], [2200], [2500], [2700], [3000], [3200], [3500]])
house_prices = np.array([[200], [240], [300], [350], [400], [420], [480], [500], [550], [580], [620]]) + np.random.randn(11, 1) * 20  # Adding slight noise

# Step 2: Split data into training and testing sets
# 80% training, 20% testing to evaluate model performance
x_train, x_test, y_train, y_test = train_test_split(house_sizes, house_prices, test_size=0.2, random_state=42)

# Step 3: Create and fit the Linear Regression model
# The model finds the best line (y = mx + c) to predict house prices from sizes
model = LinearRegression()  # Initialize the model
model.fit(x_train, y_train)  # Train the model on training data

# Step 4: Make predictions
# Predict house prices for test data
y_pred = model.predict(x_test)

# Step 5: Evaluate the model
# Mean Squared Error (MSE) measures prediction accuracy
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse:.2f} (in $1000s squared)")

# Print model parameters
# Slope indicates price increase per square foot; intercept is base price
print(f"Slope (price per sq ft): {model.coef_[0][0]:.2f} ($1000s per sq ft)")
print(f"Intercept (base price): {model.intercept_[0]:.2f} ($1000s)")

# Step 6: Plot the data and regression line
# Visualize how well the model fits the house size vs. price data
plt.figure(figsize=(10, 6))  # Set figure size
plt.scatter(x_train, y_train, color='blue', label='Training Data')  # Plot training points
plt.scatter(x_test, y_test, color='green', label='Test Data')  # Plot test points
plt.plot(house_sizes, model.predict(house_sizes), color='red', label='Regression Line')  # Plot fitted line
plt.xlabel('House Size (sq ft)')  # Label x-axis
plt.ylabel('House Price ($1000s)')  # Label y-axis
plt.title('Linear Regression: House Size vs. Price')  # Title
plt.legend()  # Show legend
plt.grid(True)  # Add grid
plt.show()  # Display the plot

# Explanation of Linear Regression:
# - Models relationship between house size (x) and price (y) as a straight line.
# - Slope shows how much price increases per square foot.
# - Intercept is the theoretical price when size is 0 (though not realistic here).
# - MSE indicates prediction error; lower is better.
# - Assumptions: Linear relationship, independent errors, constant variance.
```
